```python
import json
import requests
import pandas as pd
from requests.packages.urllib3.exceptions import InsecureRequestWarning

requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

df = pd.read_csv('/data/cnwls/cnwls_obsap/ptngu_warehouse.csv', delimiter=';' , dtype=str)


CompanyDB = "PTNGU" 
UserName =  "manager"
Password = "password"
url =  "https://192.168.250.40:50000/b1s/v1/"

mylist = df.values.tolist()

#print(mylist)

appSession = requests.Session()

#########################
# LOGIN
#########################

urllogin = url + "Login"

headers= {"B1S-ReplaceCollectionsOnPatch":"true"}
payload = { "CompanyDB" :CompanyDB ,
			"UserName" : UserName,
			"Password" : Password
			}
response = appSession.post(urllogin, json=payload,verify=False)

print(response.json())


#########################
# UPDATE UOM ORDER 
#########################




for line in mylist : 

	urlbp = url + "Warehouses('" + str(line[0]) +  "')"
	 

	rsp = appSession.delete(urlbp,verify=False)
	print(urlbp)
	#print(payload)
	print(rsp.status_code)
	if (rsp.status_code ==404) or (rsp.status_code == 400 ):
		print(urlbp)
		print(rsp.json())

	#apdp = rsp.json()


urllogout = url + "Logout"
rsplogout = appSession.post(urllogout,json=payload,verify=False)

```
