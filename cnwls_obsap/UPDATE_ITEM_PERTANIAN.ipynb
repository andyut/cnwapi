```python
import json
import requests
import pandas as pd
from requests.packages.urllib3.exceptions import InsecureRequestWarning

requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

df = pd.read_csv('/data/api/cnwapi/cnwls_obsap/udf_pertanian.csv', delimiter=';')


CompanyDB = "PTSCA" 
UserName =  "manager"
Password = "password"
url =  "https://192.168.250.40:50000/b1s/v1/"

mylist = df.values.tolist()


appSession = requests.Session()

#########################
# LOGIN
#########################

urllogin = url + "Login"


payload = { "CompanyDB" :CompanyDB ,
			"UserName" : UserName,
			"Password" : Password
			}
response = appSession.post(urllogin, json=payload,verify=False)

print(payload)
print(response.json())


#########################
# UPDATE UOM ORDER 
#########################




for line in mylist : 

	urlbp = url + "Items('"  + line[0] + "')"
   
	payload = {
				"U_pertanian" : line[2] 
			}               

	rsp = appSession.patch(urlbp,json=payload,verify=False)
	#print(urlbp)
	##print(payload)
	#print(rsp.status_code)
	if rsp.status_code >=400 :
		print(urlbp)
		print(payload)
		print(rsp.json())

	#apdp = rsp.json()


urllogout = url + "Logout"
rsplogout = appSession.post(urllogout,json=payload,verify=False)

```
